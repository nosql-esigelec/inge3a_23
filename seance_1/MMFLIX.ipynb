### Problme:
#### User Story
En tant qu'utilisateur, j'aimerais tre capable de retrouver des films par pays. Je ne veux voir que les titres des films. Je dois tre capable de spcifier les pays sous forme de liste.

#### Task

Crez une fonction get_movies_by_country qui fait une recherche par pays et retourne les titres des films trouvs.

```python
def get_movies_by_country(countries, movies):
    try:
        # here's an example of the find() query
        return list(movies.find(
            {"countries": {"$in": countries}},
            {"title": 1}
        ))
    except Exception as e:
        return e
```

```python
get_movies_by_country(["USA"], movies)
```

### Problem:
#### User Story
"As a user, I'd like to be able to search movies by cast members, genre, or perform a text search of the plot summary, full plot, and title."

#### Task
For this ticket, you will need to modify the method build_query_sort_project in db.py to allow the following movie search criteria:

**genres**: finds movies that include any of the wanted genres.
Already, the build_query_sort_project method is able to return results for two different types of movie search criteria:

**text**: performs a text search in the movies collection

**cast**: finds movies that include any of the wanted cast
You just need to construct the query that queries the movies collection by the genres field.

```python
movies.distinct('cast')
```

```python
def build_query_sort_project(filters):
    query = {}
    sort = [("tomatoes.viewer.numReviews", 1)]
    project = None
    if filters:
        if "text" in filters:
            query = {"$text": {"$search": filters["text"]}}
            meta_score = {"$meta": "textScore"}
            sort = [("score", meta_score)]
            project = {"score": meta_score}
        elif "cast" in filters:
            query = {"cast": {"$in": filters["cast"]}}
        elif "genres" in filters:
            # here's an implementation of the "genres" query
            query = {"genres": {"$in": filters["genres"]}}
    return query, sort, project
```

```python
build_query_sort_project({'cast':["Salma Hayek"], 'text': 'Cynical'})
```

## Operations on cusrsor

### Limiting

```python
limited = movies.find(
{"directors": "Sam Raimi"},
{"_id":0, "title": 1, "cast":1}
).limit(2)
print(dumps(limited, indent=2))
```

```python
pipeline = [
{"$match" : {"directors": "Sam Raimi"}},
{"$project": {"_id": 0, "title":1, "cast":1}},
{"$limit": 2}
]
```

```python
limited_agg = movies.aggregate(pipeline)
print(dumps(limited_agg, indent=2))
```

### Sorting

```python
sorted_cursor = movies.find(
    {"directors": "Sam Raimi"},
    {"_id":0, "year":1, "title":1, "cast": 1}
).sort("year", ASCENDING)
print(dumps(sorted_cursor, indent=2))
```

```python
pipeline = [
{"$match" : {"directors": "Sam Raimi"}},
{"$project": {"_id": 0, "year":1, "title":1, "cast":1}},
{"$sort": {"year": DESCENDING}}
]
```

```python
sorted_agg = movies.aggregate(pipeline)
print(dumps(sorted_agg, indent=2))
```

```python
sorted_cursor = movies.find(
{"cast": "Tom Hanks"},
{"_id": 0, "year":1, "title":1, "cast":1}
).sort([("year", DESCENDING), ("title", ASCENDING)])
print(dumps(sorted_cursor, indent=2))
```

```python
pipeline = [
{"$match": {"cast": "Tom Hanks"}},
{"$project": {"_id": 0, "year":1, "title":1, "cast":1}},
{"$sort": {"year": DESCENDING, "title": ASCENDING}}]
sorted_agg = movies.aggregate(pipeline)
print(dumps(sorted_agg, indent=2))
```

### Skipping

```python
#count the number of documents that meet those expectations
pipeline = [
    {"$match": {"directors": "Sam Raimi"}},
    {"$project": {"_id": 0, "title":1, "cast":1}},
    {"$count": "num_movies"}
]
aggregation = movies.aggregate(pipeline)
print(dumps(aggregation,indent=2))
```

```python
#skip records and take only the last one
skipped = movies.find({"directors": "Sam Raimi"},{"_id": 0, "title":1, "cast":1}).skip(10)
print(dumps(skipped))
```

```python
skipped_sorted = movies.find({"directors": "Sam Raimi"},{"_id": 0, "title":1, "cast":1}).sort("year").skip(10)
print(dumps(skipped_sorted, indent=2))
```

```python
pipeline = [
{"$match": {"directors": "Sam Raimi"}},
{"$project": {"_id":0, "year":1, "title":1,"cast":1}},
{"$sort": {"year": ASCENDING}},
{"$skip": 10}]

sorted_skipped_agg = movies.aggregate(pipeline)
print(dumps(sorted_skipped_agg, indent=2))
```
